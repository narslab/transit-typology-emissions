```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import LineString
import matplotlib.pyplot as plt
from matplotlib_scalebar.scalebar import ScaleBar 
```

```python
# Load the shapes.txt file
shapes = pd.read_csv(f"../../data/raw/networks/MTA New York City/shapes.txt")

# Group by shape_id to create LineString geometries for each route
routes = shapes.groupby("shape_id").apply(
    lambda x: LineString(zip(x["shape_pt_lon"], x["shape_pt_lat"]))
)

# Create a GeoDataFrame from the LineString geometries
routes_gdf = gpd.GeoDataFrame(
    {"shape_id": routes.index, "geometry": routes}, crs="EPSG:4326"
)

# Get bounds for the entire dataset to use the same scale
total_bounds = routes_gdf.total_bounds  # [minx, miny, maxx, maxy]
print(total_bounds)

max_width0 = total_bounds[2]-total_bounds[0]
max_height0 = total_bounds[3]-total_bounds[1]

print("max_width0", max_width0)
print("max_height0", max_height0)

# Define the color for the network
color = "#ff7f0e"

# Plot the transit network
fig, ax = plt.subplots(figsize=(12, 7))

# Plot the routes with thinner lines and specified color
routes_gdf.plot(ax=ax, color=color, linewidth=2)

# Set axis limits to the total bounds (tight rectangle)
ax.set_xlim(total_bounds[0], total_bounds[2])
ax.set_ylim(total_bounds[1], total_bounds[3])

# Remove the axis for a cleaner look
ax.axis("off")

# Save the figure
plt.savefig(r'../../figures/MTA_NYC_network.png', dpi=300, bbox_inches=None)


# Show the plot
plt.show()
```

```python
# Function to reproject data and calculate consistent bounds
def reproject_and_get_bounds(routes_gdf, target_crs="EPSG:3857"):
    # Reproject to the target CRS
    routes_gdf = routes_gdf.to_crs(target_crs)
    bounds = routes_gdf.total_bounds
    center = ((bounds[0] + bounds[2]) / 2, (bounds[1] + bounds[3]) / 2)
    width = bounds[2] - bounds[0]
    height = bounds[3] - bounds[1]
    return bounds, center, width, height

# Initialize lists to store bounds, centers, widths, and heights
all_bounds = []
all_centers = []
all_widths = []
all_heights = []

# Process each transit network

# MTA New York City
shapes_nyc = pd.read_csv("../../data/raw/networks/MTA New York City/shapes.txt")
routes_nyc = shapes_nyc.groupby("shape_id").apply(
    lambda x: LineString(zip(x["shape_pt_lon"], x["shape_pt_lat"]))
)
routes_gdf_nyc = gpd.GeoDataFrame({"geometry": routes_nyc}, crs="EPSG:4326")
bounds_nyc, center_nyc, width_nyc, height_nyc = reproject_and_get_bounds(routes_gdf_nyc)
all_bounds.append(bounds_nyc)
all_centers.append(center_nyc)
all_widths.append(width_nyc)
all_heights.append(height_nyc)

# Dallas DART
shapefile_dallas = "../../data/raw/networks/Dallas DART/Rail_Routes.shp"
routes_gdf_dallas = gpd.read_file(shapefile_dallas)
bounds_dallas, center_dallas, width_dallas, height_dallas = reproject_and_get_bounds(routes_gdf_dallas)
all_bounds.append(bounds_dallas)
all_centers.append(center_dallas)
all_widths.append(width_dallas)
all_heights.append(height_dallas)

# Cleveland RTA
shapefile_cleveland = "../../data/raw/networks/Cleveland RTA/GCRTA_Bus_Lines.shp"
routes_gdf_cleveland = gpd.read_file(shapefile_cleveland)
bounds_cleveland, center_cleveland, width_cleveland, height_cleveland = reproject_and_get_bounds(routes_gdf_cleveland)
all_bounds.append(bounds_cleveland)
all_centers.append(center_cleveland)
all_widths.append(width_cleveland)
all_heights.append(height_cleveland)

# LA Metro
shapes_la = pd.read_csv("../../data/raw/networks/LA Metro/shapes.txt")
routes_la = shapes_la.groupby("shape_id").apply(
    lambda x: LineString(zip(x["shape_pt_lon"], x["shape_pt_lat"]))
)
routes_gdf_la = gpd.GeoDataFrame({"geometry": routes_la}, crs="EPSG:4326")
bounds_la, center_la, width_la, height_la = reproject_and_get_bounds(routes_gdf_la)
all_bounds.append(bounds_la)
all_centers.append(center_la)
all_widths.append(width_la)
all_heights.append(height_la)

# St. Louis Metro
shapefile_stl = "../../data/raw/networks/ST Louis Metro/metro_routes.shp"
routes_gdf_stl = gpd.read_file(shapefile_stl)
bounds_stl, center_stl, width_stl, height_stl = reproject_and_get_bounds(routes_gdf_stl)
all_bounds.append(bounds_stl)
all_centers.append(center_stl)
all_widths.append(width_stl)
all_heights.append(height_stl)

# Find the global maximum width and height
max_width = max(all_widths)
max_height = max(all_heights)

# Function to plot networks with consistent scaling
def plot_network(routes_gdf, color, output_path, center, max_width, max_height):
    fig, ax = plt.subplots(figsize=(12, 10))
    
    # Reproject the data
    routes_gdf = routes_gdf.to_crs("EPSG:3857")
    
    # Plot the routes
    routes_gdf.plot(ax=ax, color=color, linewidth=2)
    
    # Set consistent axis limits
    x_center, y_center = center
    ax.set_xlim(x_center - max_width / 2, x_center + max_width / 2)
    ax.set_ylim(y_center - max_height / 2, y_center + max_height / 2)
    
    # Remove the axis for a clean look
    ax.axis("off")
    
    # Save the figure
    plt.savefig(output_path, dpi=300, bbox_inches="tight")
    plt.close(fig)

# Plot each network
plot_network(routes_gdf_nyc, "#ff7f0e", "../../figures/MTA_NYC_fixed_scale.png", center_nyc, max_width, max_height)
plot_network(routes_gdf_dallas, "#1f77b4", "../../figures/Dallas_DRT_fixed_scale.png", center_dallas, max_width, max_height)
plot_network(routes_gdf_cleveland, "#2ca02c", "../../figures/Cleveland_RTA_fixed_scale.png", center_cleveland, max_width, max_height)
plot_network(routes_gdf_la, "#d62728", "../../figures/LA_Metro_fixed_scale.png", center_la, max_width, max_height)
plot_network(routes_gdf_stl, "#9467bd", "../../figures/ST_Louis_Metro_fixed_scale.png", center_stl, max_width, max_height)
```

```python
# Calculate global width and height
network_widths = [bounds[2] - bounds[0] for bounds in all_bounds]  # Width of each network
network_heights = [bounds[3] - bounds[1] for bounds in all_bounds]  # Height of each network

global_width = max(network_widths)  # Global width is the maximum width among all networks
global_height = max(network_heights)  # Global height is the maximum height among all networks


def plot_network_corrected_scale(routes_gdf, color, output_path, center):
    fig, ax = plt.subplots(figsize=(10, 10))  # Keep the plot square for equal scaling
    
    # Reproject to EPSG:3857
    routes_gdf = routes_gdf.to_crs("EPSG:3857")
    
    # Plot the routes
    routes_gdf.plot(ax=ax, color=color, linewidth=2)
    
    # Set axis limits using the global width and height, centered on the network
    x_center, y_center = center
    ax.set_xlim(x_center - global_width / 2, x_center + global_width / 2)
    ax.set_ylim(y_center - global_height / 2, y_center + global_height / 2)
    
    # Add scalebar
    scalebar = ScaleBar(1, units="m", location="lower right", scale_loc="bottom", color="black")
    ax.add_artist(scalebar)
    
    # Annotate axes
    ax.set_xlabel("X Coordinates (meters)")
    ax.set_ylabel("Y Coordinates (meters)")
    
    # Save the figure
    plt.savefig(output_path, dpi=300, bbox_inches="tight")
    plt.close(fig)


# Replot all networks with corrected scaling
plot_network_corrected_scale(routes_gdf_nyc, "#ff7f0e", "../../figures/MTA_NYC_fixed_scale.png", center_nyc)
plot_network_corrected_scale(routes_gdf_dallas, "#1f77b4", "../../figures/Dallas_DRT_fixed_scale.png", center_dallas)
plot_network_corrected_scale(routes_gdf_cleveland, "#2ca02c", "../../figures/Cleveland_RTA_fixed_scale.png", center_cleveland)
plot_network_corrected_scale(routes_gdf_la, "#d62728", "../../figures/LA_Metro_fixed_scale.png", center_la)
plot_network_corrected_scale(routes_gdf_stl, "#9467bd", "../../figures/ST_Louis_Metro_fixed_scale.png", center_stl)
```
